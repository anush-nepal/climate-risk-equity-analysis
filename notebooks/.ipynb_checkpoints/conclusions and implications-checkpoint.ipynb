## Executive Summary
This analysis examined whether climate events create abnormal returns in equity markets using 8 years of daily stock data (2017-2024) from 45 companies across three climate-sensitive sectors. Contrary to common assumptions, statistical testing reveals that climate events are associated with **statistically significant positive abnormal returns** rather than negative impacts.

## Key Research Question
**Do investors price climate risk into equity valuations, and do climate events create measurable abnormal returns?**

## Methodology Summary
- **Sample**: 90,424 daily observations from 45 companies (Energy, Insurance, Real Estate)
- **Time Period**: January 2017 - December 2024
- **Climate Events**: 7 major events (hurricanes, wildfires, extreme weather)
- **Statistical Methods**: Event study methodology, t-tests, confidence intervals
- **Event Windows**: Â±10 days around each climate event

**Disclosure**: This analysis was conducted independently with all data collection, statistical testing, and interpretation performed by the author. Claude AI was used to help structure and refine the presentation of findings in this conclusions notebook.

## Key Statistical Findings

### Primary Result - Climate Events Create Positive Abnormal Returns
- **Mean return during climate events**: 0.1662% per day
- **Statistical significance**: p < 0.001 (highly significant)
- **95% Confidence interval**: [0.1139%, 0.2185%] daily returns
- **Sample size**: 4,319 event window observations

### Comparison with Normal Market Conditions
- **Event period returns**: 0.1662% daily
- **Normal period returns**: 0.0545% daily  
- **Abnormal return premium**: 0.1117% daily during events
- **Two-sample t-test**: p = 0.0005 (statistically significant)
- **Effect size**: Small but economically meaningful (Cohen's d = 0.054)

### Sector Performance Rankings
1. **Insurance**: Best risk-adjusted returns (16.35% annual return, 25.5% volatility)
2. **Energy**: Highest returns but highest risk (16.49% annual return, 36.5% volatility)  
3. **Real Estate**: Most conservative profile (11.95% annual return, 25.4% volatility)

## Economic Interpretation

### Why Climate Events Generate Positive Returns

**Supply Shock Effects (Energy Sector)**
- Extreme weather disrupts oil/gas production and refining capacity
- Reduced supply leads to higher commodity prices
- Energy companies benefit from temporary price spikes
- Markets anticipate infrastructure rebuilding demand

**Insurance Premium Adjustments**
- Post-disaster reassessment of risk leads to premium increases
- Insurance companies can often raise rates faster than claims materialize
- Regulatory approval for emergency premium adjustments
- Long-term contracts lock in higher pricing

**Economic Stimulus and Recovery Spending**
- Government disaster relief and infrastructure rebuilding
- Federal emergency declarations trigger spending programs
- Construction and materials demand surge
- Economic multiplier effects in affected regions

### Market Efficiency Considerations

**Information Processing**
- Markets may initially underprice climate event impacts
- Professional traders exploit temporary dislocations
- Algorithm-driven strategies capitalize on volatility
- Recovery patterns become more predictable over time

**Sector Rotation Effects**  
- Flight to quality within climate-exposed sectors
- Large-cap stocks benefit from flight from smaller, riskier companies
- Portfolio rebalancing creates buying pressure

## Limitations and Caveats

### Methodological Limitations

**Sample Selection Bias**
- Analysis limited to large-cap companies that survived the entire study period
- Excludes smaller companies that may be more vulnerable to climate events
- Survivorship bias may overstate positive returns

**Event Selection and Definition**
- Limited to 7 major climate events over 8 years
- Event timing based on media coverage dates, not economic impact dates
- May not capture gradual or chronic climate effects
- Selection criteria could introduce bias toward high-visibility events

**Market Condition Dependencies**
- Study period includes significant market volatility (COVID-19, energy transitions)
- Results may not generalize to different market environments
- Bull market conditions during much of study period may influence findings

### Statistical Limitations

**Event Window Methodology**
- Fixed 20-day windows may not capture full economic impact duration
- Overlapping events could contaminate individual event analysis
- Non-trading days and market closures create gaps in analysis

**Return Attribution**
- Cannot definitively isolate climate event effects from other concurrent factors
- Correlation does not prove causation
- Multiple events during same periods complicate attribution

### External Validity Concerns

**Geographic and Temporal Scope**
- Limited to US equity markets (2017-2024)
- Climate impact patterns may differ in other countries or time periods
- Regulatory and market structure differences limit global applicability

## Implications and Future Research

### Investment Implications

**Portfolio Construction**
- Climate events may create short-term trading opportunities rather than long-term risks
- Energy sector offers highest returns but requires tolerance for extreme volatility
- Insurance sector provides best risk-adjusted exposure to climate-related returns
- Traditional "climate risk" assumptions may need reassessment

**Risk Management**
- Current climate risk models may underestimate positive correlation between events and returns
- Diversification strategies should account for sector-specific climate responses
- Event-driven investment strategies may be viable in climate-sensitive sectors

### Policy and Regulatory Considerations

**Climate Risk Disclosure**
- Current regulatory frameworks assume negative climate impacts on valuations
- Evidence suggests more nuanced relationship between climate events and market performance
- Risk disclosure requirements may need to distinguish between event types and sectors

**Insurance Regulation**
- Post-disaster premium adjustments appear to benefit insurers in short term
- Regulatory review of emergency rate-setting mechanisms may be warranted
- Balance between consumer protection and insurer profitability during climate events

### Future Research Directions

## Conclusions

This analysis challenges conventional assumptions about climate risk in financial markets. Using rigorous event study methodology on 90,424 observations across 45 companies, the evidence demonstrates that climate events are associated with statistically significant positive abnormal returns rather than negative impacts.

### Final Assessment

The relationship between climate events and equity returns appears more complex than traditional risk models suggest. While climate change represents a long-term systemic risk, short-term market reactions to discrete climate events may create profitable opportunities rather than losses. This finding has important implications for how investors, regulators, and policymakers approach climate-related financial risk.

The statistical significance and economic magnitude of these results warrant further investigation across broader samples, longer time periods, and different market conditions. Understanding these dynamics is crucial as climate events become more frequent and financial markets continue to evolve their pricing of environmental risks.
